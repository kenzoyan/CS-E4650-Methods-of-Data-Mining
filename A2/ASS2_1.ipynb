```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics.pairwise import euclidean_distances
```

```python
# prepare data

ratdata= pd.read_csv("ratdataNormChecked.csv")

ratdata_ind=ratdata[["liverind", "heartind", "appind", "batind", "tailind", "ADWBind", "gonind","BMI"]]

ratdata_ind
```

```python
dis=euclidean_distances(ratdata_ind, ratdata_ind)
dis=pd.DataFrame(dis)
dis=dis.round(3)

dis
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
```

```python
#linkage{â€˜wardâ€™, â€˜completeâ€™, â€˜averageâ€™, â€˜singleâ€™}
link="single"

for numc in range(2,9):

    cluster = AgglomerativeClustering(n_clusters=numc, affinity='euclidean', linkage=link)  
    predicted_labels=cluster.fit_predict(ratdata_ind)
    # print("Predicted_labels:\n",predicted_labels)

    silhouette_avg = silhouette_score(ratdata_ind, predicted_labels)
    print("For n_clusters ={}, Linkage method={} The average silhouette_score is :{}".format(numc,link,silhouette_avg))
```

```python
pca=PCA(n_components=1)
pca.fit(ratdata_ind)
# print(pca.explained_variance_ratio_)
data_pca=pca.fit_transform(ratdata_ind)
# data_pca
```

```python
# Test in Data after PCA

#linkage{â€˜wardâ€™, â€˜completeâ€™, â€˜averageâ€™, â€˜singleâ€™}
link="ward"

for numc in range(2,9):

    cluster = AgglomerativeClustering(n_clusters=numc, affinity='euclidean', linkage=link)  
    predicted_labels=cluster.fit_predict(data_pca)
    # print("Predicted_labels:\n",predicted_labels)

    silhouette_avg = silhouette_score(data_pca, predicted_labels)
    print("For n_clusters ={}, Linkage method={} The average silhouette_score is :{}".format(numc,link,silhouette_avg))
```
